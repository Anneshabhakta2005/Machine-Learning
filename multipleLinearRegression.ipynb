```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/MLPractice/EmployeeData.csv")
df
```

```python
df.info()
```

```python
x=df.iloc[:,0:4].values
x
```

```python
from sklearn.preprocessing import LabelEncoder

en = LabelEncoder()
x[:,0]=en.fit_transform(x[:,0])
x
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer([("oh",OneHotEncoder(),[0])],remainder="passthrough")
x=ct.fit_transform(x)
x
```

```python
y=df.iloc[:,4].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.as_integer_ratio,random_state=1)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
mse=mean_squared_error(y_test,y_pred)
rmse=mse**.5
print("Root mean square error",rmse)
mae=mean_absolute_error(y_test,y_pred)
print("Mean absolute error",mae)
```
