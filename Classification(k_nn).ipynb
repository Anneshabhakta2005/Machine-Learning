```python
import pandas as pd
df=pd.read_csv("/content/mount/MyDrive/MLPractice/Copy of Social_Networks_ads.csv")
df
```

```python
df.info()
```

```python
x=df.iloc[:,1:3]
x
```

```python
y=df.iloc[:,4].values
y
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Re-initialize x from the original DataFrame to ensure it's a DataFrame
x = df.iloc[:, 1:3].copy() # Use .copy() to prevent SettingWithCopyWarning

le = LabelEncoder()
mms = MinMaxScaler()

x.iloc[:, 0] = le.fit_transform(x.iloc[:, 0])
x = mms.fit_transform(x)

x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=0)
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
pr=knn.predict(x_test)
pr
```

```python
score=knn.score(x_test,y_test)
score
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test,pr)
cm
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)
```

```python
from sklearn.metrics import precision_score, accuracy_score,recall_score
ac=accuracy_score(y_test,pr)
pre=precision_score(y_test,pr)
rec=recall_score(y_test,pr)
print(f"Accuracy Score: {ac},Precision: {pre}, Recall: {rec}")
```

```python
import pandas as pd
df=pd.read_csv("/content/mount/MyDrive/MLPractice/Copy of CarPrice_Assignment.csv")
df
```

```python
p=df.iloc[:,1:3]
p
```

```python
q=df.iloc[:,4].values
q
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Re-initialize p from the original DataFrame to ensure it's a DataFrame
p = df.iloc[:, 1:3].copy() # Use .copy() to prevent SettingWithCopyWarning

le = LabelEncoder()
mms = MinMaxScaler()

# Encode the 'symboling' column (index 0). While already numeric,
# LabelEncoder converts it to a range of integers.
p.iloc[:, 0] = le.fit_transform(p.iloc[:, 0])

# Encode the 'CarName' column (index 1) as it contains string values
p.iloc[:, 1] = le.fit_transform(p.iloc[:, 1])

p = mms.fit_transform(p)
p
```

```python
from sklearn.model_selection import train_test_split
p_train,p_test,q_train,q_test=train_test_split(p,q,test_size=.2,random_state=0)
p_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(p_train,q_train)
```

```python
pr=knn.predict(p_test)
pr
```

```python
score=knn.score(p_test,q_test)
score
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(q_test,pr)
cm
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)
```
