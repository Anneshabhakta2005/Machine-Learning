```python
import numpy as np
a=np.arange(12)

print(a)
print("Last Element:",a[-1])
print("Dimension of a",a.ndim)
print("Shape of a:",a.shape)
print("Data typeof each element:",a.dtype)
```

```python
print(a)
print("####################")
b=a.reshape(3,4)
print(b)
print("dimension of b:",b.ndim)
print("Sum of all:",b.sum())
print("Average of all:",b.mean())
print("Row Sum:",b.sum(axis=1))
print("Column sum:",b.sum(axis=0))
```

```python
print(b)
print("################")
print(np.ravel(b))
```
